```python
import pandas as pd

df = pd.read_csv("../data/outputs/review_sentiment_themes_cleaned.csv")

# Remove duplicate reviews by review_id
df = df.drop_duplicates(subset='review_id')

# Optional: reset index
df = df.reset_index(drop=True)

# Save cleaned version
df.to_csv("../data/outputs/review_sentiment_themes.csv", index=False)
print(f"✅ Cleaned {len(df)} reviews saved.")
```
